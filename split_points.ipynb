```python
import pandas as pd
import json
pd.set_option("display.max_columns", None)

csv_data = pd.read_csv("geocode_met_data.csv", dtype = str)
csv_data
```

```python
keys = [c for c in csv_data if c.startswith('origin_')]
merged_origins = pd.melt(csv_data, id_vars='Object ID', value_vars=keys, value_name='origin')

merged_origins = merged_origins[merged_origins['origin'].notna()]
merged_origins
```

```python
joined_df = merged_origins.merge(csv_data, how='left')
joined_df
```

```python
joined_df.loc[joined_df['variable'] == 'origin_0', 'lat'] = joined_df.loc[joined_df['variable'] == 'origin_0', 'lat_0']
joined_df.loc[joined_df['variable'] == 'origin_0', 'long'] = joined_df.loc[joined_df['variable'] == 'origin_0', 'long_0']

joined_df.loc[joined_df['variable'] == 'origin_1', 'lat'] = joined_df.loc[joined_df['variable'] == 'origin_1', 'lat_1']
joined_df.loc[joined_df['variable'] == 'origin_1', 'long'] = joined_df.loc[joined_df['variable'] == 'origin_1', 'long_1']

joined_df.loc[joined_df['variable'] == 'origin_2', 'lat'] = joined_df.loc[joined_df['variable'] == 'origin_2', 'lat_2']
joined_df.loc[joined_df['variable'] == 'origin_2', 'long'] = joined_df.loc[joined_df['variable'] == 'origin_2', 'long_2']

joined_df.loc[joined_df['variable'] == 'origin_3', 'lat'] = joined_df.loc[joined_df['variable'] == 'origin_3', 'lat_3']
joined_df.loc[joined_df['variable'] == 'origin_3', 'long'] = joined_df.loc[joined_df['variable'] == 'origin_3', 'long_3']

joined_df
```

```python
joined_df.drop(['variable', "Unnamed: 0.1",	"Unnamed: 0", 'origin_0', 'origin_1', 'origin_2', 'origin_3', 'lat_0', 'long_0', 'lat_1', 'long_1', 'lat_2', 'long_2', 'lat_3', 'long_3'], axis=1, inplace=True)
joined_df
```

```python
joined_df.to_csv('final_met_data.csv')
```
